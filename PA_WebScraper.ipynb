## Scraper

```python
def getHomePage(HOME_URL):
    """Recibe el enlace de la página y hace la consulta"""
    
    try:
        response = requests.get(HOME_URL) # Consulta a la página a scrapear
        if response.status_code == 200:
            home_soup = BeautifulSoup(response.content, 'lxml')
            return home_soup
        
    except Exception as e:
        print(e)
```

```python
def getNavSections(home_soup):
    """Recibe la sopa de la página principal, retorna los enlaces de las secciones"""
    
    try:
        nav_soup = home_soup.find('ul', attrs={'class':'main-menu-links-2019'}).find_all('li')[:-3]
        navs = [sec.a.get('href') for sec in nav_soup]
        return navs[2] 
    
    except Exception as e:
        print(e)  
   
```

```python
def getSectionNewsLinks(HOME_URL, economia_section):
    """Recibe los enlaces de las secciones, retorna los enlaces de las noticias en ellas """
    
    economia_blog = HOME_URL + economia_section
    section_response = requests.get(economia_blog) # Consulta las secciones
    economia_news_url = []    
    try:
        if section_response.status_code == 200:
            articles = economia_soup.find('div', attrs={'class':'view-footer'}).find_all('article')
            economia_articles_url = [article.a.get('href') for article in articles]
            economia_news_url.append(economia_articles_url)
                
    except Exception as e:
        print(e)   
            
    return economia_news_url
```

```python
def getInfo(HOME_URL, economia_news_url): 
        
    try:
        for list_articles in economia_news_url:
            info = []
            for articles in list_articles:
                article_url = HOME_URL + articles
                
                article_response = requests.get(article_url)
                
                if article_response.status_code == 200:
                    article_soup = BeautifulSoup(article_response.content, 'lxml')
                    
                    # Título de la noticia
                    title = article_soup.find('h1').get_text()
                   
                    # Fecha de publicación
                    date = article_soup.find('li', attrs={'class':'fecha-hora'}).get_text()
                    date = re.findall(r'[0-9]*/[0-9]*/[0-9]{2,4}', date)
                    
                    #Resumen de la noticia
                    summary = article_soup.find('p', attrs={'class':'subheader'}).get_text()
                    
                    # Cuerpo de la noticia
                    body = article_soup.find_all('p')[3:-6]
                    body = [pr.get_text() for pr in body]
                    
                    info.append({'title': title, 
                                 'date': date,
                                 'summary': summary,
                                 'body' : body})
                    
    except Exception as e:
                print(e)
                    
    return info
```

```python
def runScraper(HOME_URL):
    home = getHomePage(HOME_URL)
    navs = getNavSections(home_soup)
    urls = getSectionNewsLinks(HOME_URL, navs)
    info = getInfo(HOME_URL, urls)
    
    return info
```

```python
HOME_URL = 'https://www.panamaamerica.com.pa'

raspado = runScraper(HOME_URL)
```

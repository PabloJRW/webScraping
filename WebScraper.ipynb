```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://pagina12.com.ar'
```

```python
response = requests.get(url)
response.status_code
```

```python
response.content
```

```python
s = BeautifulSoup(response.content, 'lxml')
type(s)
```

```python
print(s.prettify())
```

```python
secciones = s.find('ul', attrs={'class':'horizontal-list main-sections hide-on-dropdown'}).find_all('li')
secciones
```

```python
links_secciones = [seccion.a.get('href') for seccion in secciones]
links_secciones
```

```python
economia = requests.get(links_secciones[1])
economia
```

```python
s_economia = BeautifulSoup(economia.content, 'lxml')
print(s_economia.prettify())
```

```python
economia_page = s_economia.find(id='main-content').find_all('article')
economia_page
```

```python
economia_articles = [article.a.get('href') for article in economia_page]
economia_articles
```

```python
url_news = economia_articles[4]
url_news
```

```python
try:
    url_to_news = url + url_news # Enlace a la seccin -> [deportes, economa, sociedad, etc.]
    r = requests.get(url_to_news) # Consulta https
    if r.status_code == 200:
        sect = BeautifulSoup(r.content, 'lxml') #Seccion -> [deportes, economa, sociedad, etc.]
        title = sect.find('div', attrs={'class':'col 2-col'}).find('h1').get_text()  # Ttulo de la noticia
        raw_content = sect.find('div', attrs={'class':'article-main-content article-text'}).find_all('p') # Contenido
        content = [p.get_text() for p in raw_content] # Contenido
        print(title)
        print(content)

except Exception as e:
    print(e)
```
